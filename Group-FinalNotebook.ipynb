# The Best Neighborhood in Pittsburgh
For our final project, our group tried to figure out the best way to determine which Pittsburgh neighborhood can be considered the "best" using datasets from WPRDC. Since "best" is very subjective, we decided to find metrics that we thought would fit together well and work hand in hand. We decided on park/open space availability, police incident counts, and 311 service data. We chose these because they reflect overall activity/liveability (parks), safety (police incidents), and general conditions (311 data). We found these by going to WPRDC and looking through the available datasets. Some alternatives we also thought about exploring were employment rates/opportunities, amount of businesses, and poverty rates, as those would also represent economic stability.

# Dataset 1: 311 Data
For our first dataset, we analyzed the number of 311 service requests made in each Pittsburgh neighborhood. These requests include issues that are non-emergencies, such as debris, potholes, noise complaints, general maintenance and more. The dataset contained information about where the service was required, what for, and when it occurred, however we just looked at the number of requests. Lower levels suggest that fewer problems were occuurring and requires less upkeep. Some neighborhoods had very low levels of issues, while some had much higher counts. Troy Hill ended up in a middle range, not being too high but also not being very low. Troy Hill had the 2nd least amount of requests of the 7 neighborhoods at the average, being 7,938 requests.

# Dataset 2: Criminal Incidents
Our second dataset looked at the number of police incidetns reported in each neighborhood. Lower numbers of reports represent safer areas with fewer police-reported events. Many neighborhoods had a low count such as Ridgemont, Mount Oliver, and Swisshelm Park, while more active neighborhoods like South Side Flats, Carrick, and Shadyside had the highest numbers. Troy Hill fell into the middle part of the distribution, therefore it isn't the safest overall but it also avoids very high levels of crime/police activity, suggesting that it is relatively safe compared to many of the other neighborhoods.

# Dataset 3: Parks & Open Spaces
The third dataset focused on the number of parks and open spaces found in each neighborhood. After filtering the dataset to only include said park and open space entries and removing parks with missing neighborhoods, we counted how many each area had. Most neighborhoods had 1-3 parks, while only a few reached 4+. Troy Hill performed very well here, containing 5 parks and placing it in the top 6. This gave Troy Hill a very strong advantage in quality of life and recreation.

# Conclusion:
After reviewing all three datasets, being parks/open spaces, police incidents, and 311 data, we found that **Troy Hill** stood out as the most balanced overall neighborhood. It has one of the highest counts of parks and open spaces, making it strong in recreational activities. It also was on the lower end of the average 311 complaints and police incident totals, showing it was safer than average and well managed. While other neighborhoods performed well in one category but poorly in another, **Troy Hill** performed well across all 3, which is why we chose it as the best neighborhood.
