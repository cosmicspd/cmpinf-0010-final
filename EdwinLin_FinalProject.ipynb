```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
CR = pd.read_csv("Criminal_record.csv", index_col="_id")
CR["INCIDENTNEIGHBORHOOD"] = (
    CR["INCIDENTNEIGHBORHOOD"]
    .str.title()
    .str.strip()
)

crime_counts = CR["INCIDENTNEIGHBORHOOD"].value_counts()
community_dict = {name: int(count) for name, count in crime_counts.items()}
community_dict #crime count in each community
```

##Sort neighborhoods from lowest crime count to highest

##The neighborhood with the fewest crimes = safest = best

```python
sorted_crime = crime_counts.sort_values()
sorted_crime.head(10)

```

```python
top10_safest = sorted_crime.head(10)

plt.figure(figsize=(12,6))
plt.bar(top10_safest.index, top10_safest.values, color='green')
plt.xticks(rotation=45, ha='right')
plt.ylabel("Crime Count")
plt.title("Top 10 Safest Neighborhoods (Lowest Crime Count)")
plt.tight_layout()
plt.show()
```

# BUT

```python
top10_dangerous = sorted_crime.tail(10)

plt.figure(figsize=(12,6))
plt.bar(top10_dangerous.index, top10_dangerous.values, color='red')
plt.xticks(rotation=45, ha='right')
plt.ylabel("Crime Count")
plt.title("Top 10 Neighborhoods With the Most Crimes")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14,6))
plt.bar(sorted_crime.index, sorted_crime.values, color='skyblue')
plt.xticks(rotation=90)
plt.ylabel("Crime Count")
plt.title("Crime Count by Neighborhood (All Pittsburgh Neighborhoods)")
plt.tight_layout()
plt.show()
```
